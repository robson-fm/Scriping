```python
# Importa as bibliotecas necessárias para o funcionamento do script.
from selenium import webdriver  # Permite controlar um navegador web
from selenium.webdriver.common.by import By  # Utilizado para localizar elementos dentro de uma página
from selenium.webdriver.chrome.options import Options  # Permite configurar opções para o navegador Chrome
from selenium.webdriver.chrome.service import Service  # Serviço do Chrome WebDriver
from webdriver_manager.chrome import ChromeDriverManager  # Gerencia a instalação/atualização do ChromeDriver
import json  # Biblioteca para trabalhar com dados no formato JSON
import pandas as pd
```

```python
options = Options()
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'
options.add_argument('--headless')
#options.add_argument('--disable')
options.add_argument(f'user-agent={user_agent}')
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
driver.get('https://www.olx.com.br/estado-sp?q=apartamento')
```

```python
dados = driver.find_element(By.ID, '__NEXT_DATA__').get_attribute('innerHTML')
```

```python
data = json.loads(dados)
```

```python
data['props']['pageProps']['ads'] [0]
#title
#preco
#location
#url
```

```python
ads_data =[]
```

```python
for ad in data['props']['pageProps']['ads']:
    title = ad.get('title' , '-')
    preco = ad.get('price', '-')
    location = ad.get('location', '-')
    url= ad.get ('url', '-')
    ads_data.append({'Título': title, 'Preço': preco, 'Localização': location, 'URL': url})
```

```python
ads_data
```

```python
df_ads = pd.DataFrame(ads_data)
```

```python
df_ads.to_excel('anuncios imoveis.xlsx', index=False, engine='openpyxl')
```

```python
driver.quit
```
